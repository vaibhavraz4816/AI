```python
# Q3. Load the data from wine dataset. Check whether all attributes are standardized or not (mean
# is 0 and standard deviation is 1). If not, standardize the attributes. Do the same with Iris dataset.
```

```python
#Standardization For Wine Dataset

import pandas as pd
import numpy as np
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
data = pd.DataFrame(wine.data ,columns = wine.feature_names)
data['Class'] = wine.target
```

```python
data
```

```python
(data.isnull() | data.empty | data.isna()).sum()
```

```python
data.mean()
```

```python
data.std()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_wine_data = pd.DataFrame(scaler.fit_transform(data.loc[:,data.columns != 'Class']),columns = data.columns[:-1])
```

```python
scaled_wine_data
```

```python
scaled_wine_data.mean()
```

```python
scaled_wine_data.std()
```
