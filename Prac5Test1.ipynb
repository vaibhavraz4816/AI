```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
#cleaning
(df.isna()|df.empty|df.isnull()).sum()
```

```python
df = df.rename(columns = {'Outcome' : 'target'})
```

```python
df
```

```python
df['target'].value_counts()
```

```python
df['target'] = df['target'].map({ 0 : 'No', 1 : 'Yes'})
```

```python
df.head()
```

```python
df['target'] = df['target'].map({'Yes' : 1, 'No' : 0})
```

```python
# print("Uniques values in column1",df['Pregnancies'].unique(),"\n")
# print("Uniques values in column2",df['Glucose'].unique(),"\n")
# print("Uniques values in column3",df['BloodPressure'].unique(),"\n")
# print("Uniques values in column4",df['SkinThickness'].unique(),"\n")
# print("Uniques values in column5",df['Insulin'].unique(),"\n")
# print("Uniques values in column6",df['BMI'].unique(),"\n")
# print("Uniques values in column7",df['DiabetesPedigreeFunction'].unique(),"\n")
# print("Uniques values in column8",df['Age'].unique(),"\n")
df.info()
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df2 = pd.DataFrame(scaler.fit_transform(df.loc[:,df.columns != 'target']),columns = df.columns[:-1])
df2['target'] = df['target']
```

```python
df2.describe()
```

```python
# features
X = df2.drop('target',axis = 1)
# target
y = df['target']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 42)
```

```python
print("Shape of X train : ",X_train.shape)
print("Shape of X test : ",X_test.shape)
print("Shape of Y train : ",y_train.shape)
print("Shape of Y test : ",y_test.shape)
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
deci_tree = DecisionTreeClassifier(criterion = "gini")
deci_tree.fit(X_train,y_train)
y_pred = deci_tree.predict(X_test)
```

```python
#Classification Metrics
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score,ConfusionMatrixDisplay
fig,(ax1) = plt.subplots()
fig.set_size_inches(15,5)
ax1.set_title('Confusion Matrix')
ConfusionMatrixDisplay.from_estimator(deci_tree,X_test,y_test,ax = ax1)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print("Accuracy of decision tree")
deci_score = accuracy_score(y_test,y_pred)
print("Decision tree accuracy : ",deci_score * 100 ,"%")

print("\n Classification report on the test data")
print(classification_report(y_test,y_pred))
```

```python
from sklearn import tree
fn = list(df2.columns[:-1])
cn = ['No','Yes']

fig , axes = plt.subplots(nrows = 1,ncols = 1 ,figsize = (8,8) ,dpi = 400)
tree.plot_tree (deci_tree ,feature_names = fn,class_names = cn,filled = True);
fig.savefig('MyDecitree.png')
```

```python
#naive bayse in holdout
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,y_train)
```

```python
y_pred = nb.predict(X_test)
```

```python
fig , axes = plt.subplots(figsize = (15,5))
axes.set_title("Confusion Matrix")
ConfusionMatrixDisplay.from_estimator(nb,X_test,y_test,ax = axes)
```

```python
print("Accuracy of Naive Bayes on test data")
nb_score = accuracy_score(y_test,y_pred)
print("Naive bayes accuracy : ",nb_score*100 ,"%")

print("\n Classification report on test data")
print(classification_report(y_test,y_pred))
```

```python
#knn in holdout
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
```

```python
print("Accuracy of K Nearest Neighbours on test data")
knn_score = accuracy_score(y_test,y_pred)
print("Naive bayes accuracy : ",knn_score*100 ,"%")

print("\n Classification report on test data")
print(classification_report(y_test,y_pred))
```

```python
#random subsampling
#decision tree
acc = []
deci_tree = DecisionTreeClassifier(criterion = "entropy")
for i in range(10):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = i+1)
    deci_tree.fit(X_train,y_train)
    y_pred = deci_tree.predict(X_test)
    acc.append(accuracy_score(y_test,y_pred))
deci_random = sum(acc)/len(acc)
deci_random
```

```python
#random subsampling
#naive bayse
acc_naive = []
nb_random = GaussianNB()
for i in range(10):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = i+1)
    nb_random.fit(X_train,y_train)
    y_pred = nb_random.predict(X_test)
    acc_naive.append(accuracy_score(y_test,y_pred))
nb_random = sum(acc_naive)/len(acc_naive)
nb_random
```

```python
#random subsampling
#knn
acc_knn = []
knn_random = KNeighborsClassifier()
for i in range(10):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = i+1)
    knn_random.fit(X_train,y_train)
    y_pred = knn_random.predict(X_test)
    acc_knn.append(accuracy_score(y_test,y_pred))
knn_random = sum(acc_knn)/len(acc_knn)
knn_random
```

```python
#Cross validation
from sklearn.model_selection import KFold,StratifiedKFold,LeaveOneOut
from sklearn.model_selection import cross_val_score
```

```python
#Cross validation
#Decision Tree
kf = KFold(n_splits = 10)
deci_cross = cross_val_score(deci_tree,X,y,cv = kf)
deci_cross = deci_cross.mean()
print("Accuracy of decision tree on cross validation : ",deci_cross*100)
```

```python
#Cross validation
#Naive Bayse
loo = LeaveOneOut()
nb_cross = cross_val_score(nb,X,y,cv = loo)
nb_cross = deci_cross.mean()
print("Accuracy of decision tree on cross validation : ",nb_cross*100)
```

```python
#Cross validation
#knn
skf = StratifiedKFold(n_splits = 10)
knn_cross = cross_val_score(knn,X,y,cv = skf)
knn_cross = knn_cross.mean()
print("Accuracy of decision tree on cross validation : ",knn_cross*100)
```
